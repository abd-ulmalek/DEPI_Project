```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\cm\Desktop\DEPI Project\electronics_supplyChain.csv")
```

```python
print("===== Info =====")
df.info()
```

```python

print("===== Numerical Summary =====")
df.describe()
```

```python
print("===== Missing Values =====")
print(df.isnull().sum())
```

```python
print("Duplicates:", df.duplicated().sum())

print("Duplicates Order ID:", df['Order ID'].duplicated().sum())

display(df[df.duplicated()])

df = df.drop_duplicates()

print("After cleaning:", df.duplicated().sum())
```

```python
# ---------- Dates ----------
Dates = ['Order Date', 'Ship Date', 'Delivery Date']
for date in Dates :
    df[date] = pd.to_datetime(df[date], errors='coerce', dayfirst=True)

print(df[Dates].info())
print(df[Dates].isnull().sum())
```

```python
# Keep only logically correct dates
df = df[
    (df["Ship Date"] >= df["Order Date"]) &
    (df["Delivery Date"] >= df["Ship Date"])
]
```

```python
# ---------- Product ----------
df['Product type'] = df['Product type'].str.lower().str.strip()
df['Category'] = df['Category'].str.lower().str.strip()
df['SKU'] = df['SKU'].astype(str).str.strip()
```

```python
df['Return Reason'] = df['Return Reason'].str.lower().str.strip()

df[df['Return Reason'].isnull()][['Units Returned', 'Return Reason']]

df[df['Units Returned'] == 0][['Units Returned', 'Return Reason']]

df.loc[df['Units Returned'] == 0, 'Return Reason'] = "no return"

print('Missing Values:', df['Return Reason'].isnull().sum())

df[df['Return Reason'] == "no return"]
```

```python
# ---------- Customer ----------
cols = ['Customer demographics', 'Sales Channel', 'Payment Method', 'Region']
for col in cols:
    df[col] = df[col].str.lower().str.strip() 
    
df['Customer demographics'].value_counts()
    
df['Customer demographics'] = (df['Customer demographics'].replace('unknown', np.nan).ffill())

print(df[df['Customer demographics'] == 'unknown'])

df['Customer demographics'].value_counts()

# ---------- Supplier ----------
df['Supplier name'] = df['Supplier name'].str.lower().str.strip()
```

```python
# Remove Invalid Records
df = df[
    (df["Shipping times"] >= 0) &
    (df["Price"] >= 0) &
    (df["Discount (%)"].between(0,100)) &
    (df["Number of products sold"] > 0) &
    (df["Units Returned"] <= df["Number of products sold"])
]
```

```python
# =====================================
# Final Data Quality Check
# =====================================

print("\n===== Missing Values =====")
print(df.isnull().sum())

print("\n===== Negative Prices =====")
display(df[df["Price"] < 0])

print("\n===== Invalid Returns =====")
display(df[df["Units Returned"] > df["Number of products sold"]])

print("\n===== Orders =====")
print("Unique Order IDs :", df["Order ID"].nunique())
print("Total Rows :", len(df))

print("\n===== Revenue Sample =====")
display(df[
    [
        "Price",
        "Number of products sold",
        "Discount (%)",
        "Revenue generated"
    ]
].head())
```

```python
# Final Clean Dataset
df_clean = df.copy()

print("Data is ready for analysis.")
```

```python
# =====================================
# Q9: Profit Margin & Units Sold by Product Type
# =====================================

product_profit = (
    df_clean.groupby("Product type")
    .agg({
        "Number of products sold":"sum",
        "Revenue generated":"sum",
        "Profit":"sum"
    })
)

product_profit["Profit Margin (%)"] = (
    product_profit["Profit"] /
    product_profit["Revenue generated"] * 100
)

product_profit = (
    product_profit[["Number of products sold","Profit Margin (%)"]]
    .round(2)
    .sort_values(by="Number of products sold", ascending=False)
    .reset_index()
)

product_profit
```
